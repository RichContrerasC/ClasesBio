Simulando que tiene dos puertas una m치s peque침a que otra.

```python
from pylab import *
%matplotlib inline
```

```python
def espectro (tiempo,corriente,alpha_o,alpha_c,beta_o,beta_c):
    valor = zeros((tiempo,2))
    
    valor [0,:] = [0,0]
    for i in arange (tiempo-1):
        aleatorio = rand()
        if valor [i,0] == 0:
            if aleatorio < alpha_o:
                valor [i+1,:] = [1,corriente]
            else:
                valor [i+1,:] = [0,0]
        elif valor [i,0] == 1:
            if aleatorio < alpha_c:
                valor [i+1,:] = [0,0]
            
            elif alpha_c < aleatorio < alpha_c + beta_c:
                valor [i+1,:] = [2,0]
            else:
                valor [i+1,:] = [1,corriente]
        elif valor [i,0] == 2:
            if aleatorio < beta_o:
                valor [i+1,:] = [1,corriente]
            else:
                valor [i+1,:] = [2,0]
    return valor
                
#elif se usa cuando hay m치s de dos condiciones y else solamente cuando hay 2 condiciones. 
```

```python
valor1 = espectro (300,1,0.07,0.07,0.5,0.5)
i = arange(len(valor1[:,0]))
plot (i,valor1[:,1])
ylim (-0.5,1.5)
```
