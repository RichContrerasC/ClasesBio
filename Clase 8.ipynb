# Poblaciones y caos
Ecuación de Verhulst

$$
\frac{dP}{dt}  = \mu P(1 - \frac{P}{K})
$$

donde $\mu$ es la razón de crecimiento y K es la capacidad de carga.
La ecuación logística es la aproximación discreta de la ecuación de Verhulst

$$
f(x) = \mu x(1-x)
$$

Entonces encontraremos una serie de valores para la población donde 

$$
x_{n+1} = \mu x_n(1-x_n)
$$

$$
x^* = 1-\frac{1}{\mu}
$$

```python
from pylab import *
%matplotlib inline
```

```python
def logistica (mu, xi):
    return mu*xi*(1-xi)
```

```python
logistica (2,0.9)
```

```python
def iteracion (t,mu,xi):
    poblacion = zeros(t)
    poblacion[0] = xi
    for i in arange (t-1):
        poblacion [i+1] = logistica(mu, poblacion[i])
        
    return poblacion
```

```python
def evolucion(t,mu,xi):
    poblacion = iteracion(t,mu,xi)
    x = arange (len(poblacion))
    plot (x,poblacion[x],'o')
```

```python
for i in arange (20):
    evolucion(400,0 + 0.2*i,0.45)
    ylim (0,1)
    print 0.2*i
```
