# Clase 4 Ecuaciones diferenciales de segundo orden
El orden de las ecuaciones nos permiten delimitar los grados de libertad de la ecuación, primer orden es un grado, segundo orden dos grados.

```python
from pylab import *
%matplotlib inline
```

Resolveremos

$$
y''= -y
$$

$$
y = [sin (x), cos(x)]
$$

$$
y´[0] = y[1]
$$

$$
y´ [1] = -y[0]  
$$

```python
def segundo_orden(y,t):
    return array([y[1],-y[0]])
```

```python
def euler(func,p_ini,t_final,h):
    pasos = arange(p_ini[0],t_final+h,h)
    y = [0]*(len(pasos))
    y[0]= p_ini[1]
    for i in range(len(pasos)-1):
        y[i+1]= y[i] + func(y[i], pasos[i])*h
        sol= pasos, y
    return sol
```

```python
sol = euler(segundo_orden,[0,[0,1]],200,0.01)
y = zeros(len(sol[0]))

for i in range(len(sol[1])):
    y[i] = sol[1][i][1]
    
plot(sol[0], y)
plot(sol[0], cos(sol[0]))
```
